```python
df = pd.read_csv('congregations2.csv')
df['wlat'] = df['Members'] * df['latitude']
df['wlon'] = df['Members'] * df['longitude']
df = df.dropna(subset=['Members'])
```

```python
def average_coordinates(df):
    lat = sum(df['wlat']) / sum(df['Members'])
    lon = sum(df['wlon']) / sum(df['Members'])
    return[lat,lon]
```
